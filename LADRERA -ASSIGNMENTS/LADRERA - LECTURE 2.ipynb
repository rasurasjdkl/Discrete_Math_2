#### $\textbf{Task 1:}$ Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

---

$\textbf{Question 1: }$ *Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$*

***SOLUTION:***

**Step 1:** 

Recall the expected value formula for the multinomial distribution. For a multinomial distribution with parameters $n$ and $p_1, p_2, \ldots, p_k$:

$E[Y_i] = n \cdot p_i$

where $Y_i$ represents the number of times the $i$-th outcome occurs in $n$ trials.

**Step 2:** 

The probability mass function of the multinomial distribution is given by:

$P(y_1, y_2, \ldots, y_k) = \frac{n!}{y_1! y_2! \cdots y_k!} p_1^{y_1} p_2^{y_2} \cdots p_k^{y_k}$

where $y_i \geq 0$ and $\sum_{i=1}^k y_i = n$.

**Step 3:** 

The expected value of $ Y_i $ is calculated as:

$E[Y_i] = \sum_{y_1, y_2, \ldots, y_k} y_i \cdot P(y_1, y_2, \ldots, y_k)$

**Step 4:** 

Using the properties of the multinomial distribution, we know:

$E[Y_i] = n \cdot p_i$

**CONCLUSION**

Hence, we have proved that the expected value $E[Y_i] = n \cdot p_i$.

#### $\textbf{Task 2:}$ Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

---

$\textbf{Question 1: }$ *From the following steps shown above, derive the probability distribution of the experiment.*

***SOLUTION:***

**Step 1:** Define the probability mass function of the binomial distribution:

$P(Y = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$

where $k$ represents the number of successes observed during the $ n $ trials, $\binom{n}{k}$ is the binomial coefficient, and $p$ is the probability of success on a single trial.

**Step 2:** Substituting the given values into the probability mass function, we have:

$P(Y = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$

**Step 3:** Therefore, the probability distribution of the experiment is given by the binomial distribution formula:

$P(Y = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$

This probability distribution describes the likelihood of observing $k$ successes in $n$ trials, given the probability of success $p$ on a single trial.

---

$\textbf{Question 2: }$ *Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.*

***SOLUTION:***

**Step 1:**

The expectation, denoted as $E[Y]$, represents the average value of the random variable $Y$ over many trials. It's calculated as the sum of each possible outcome $y$ multiplied by its probability $P(y)$.

$E[Y] = \sum_{y=0}^{n} y \cdot P(y)$

**Step 2:**

In this experiment, $Y$ represents the number of successes observed during $n$ trials, where the probability of success in each trial is $p$. The probability distribution of $Y$ follows a binomial distribution. Hence, we use the binomial probability mass function:

$P(y) = \binom{n}{y} \cdot p^y \cdot (1 - p)^{n - y}$

**Step 3:**

Substitute the binomial probability mass function into the expectation formula:

$E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1 - p)^{n - y}$

**Step 4:**

Factor out the constant $p$ from the summation:

$E[Y] = np \sum_{y=0}^{n} y \cdot \frac{(n-1)!}{(y-1)!(n-y)!} \cdot p^{y-1} \cdot (1 - p)^{n - y}$

**Step 5:**

Rearrange the terms to facilitate the application of the binomial theorem:

$E[Y] = np \sum_{y=1}^{n} y \cdot \binom{n-1}{y-1} \cdot p^{y-1} \cdot (1 - p)^{n - y}$

**Step 6:**

Using the binomial theorem, we simplify the expression:

$E[Y] = np \cdot \sum_{k=0}^{n-1} \binom{n-1}{k} \cdot p^k \cdot (1 - p)^{(n-1) - k}$

$E[Y] = np \cdot 1$

$E[Y] = np$

**CONCLUSION**

Hence, the expectation of a binomial distribution, $E[Y] = np$, confirms that the mean equals the product of the number of trials and the probability of success.

**Step 1:**

The binomial probability mass function is given by:

$P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k}$

For $k = 9$:

$P(Y = 9) = \binom{10}{9} (0.30)^9 (0.70)^{1}$

$\binom{10}{9} = \frac{10!}{9!1!} = 10$

Computing the probabilities:

$(0.30)^9 \approx 0.000019683$

$(0.70)^1 = 0.7$

Then, calculate $P(Y = 9)$:

$P(Y = 9) = 10 \times 0.000019683 \times 0.7 \approx 0.000137781$

**Step 2:**

For$ k = 10$:

$P(Y = 10) = \binom{10}{10} (0.30)^{10} (0.70)^{0}$

Calculate the binomial coefficient:

$\binom{10}{10} = \frac{10!}{10!0!} = 1$

Compute the probabilities:

$(0.30)^{10} \approx 0.0000059049$

$(0.70)^0 = 1$

Then, calculate $P(Y = 10)$:

$P(Y = 10) = 1 \times 0.0000059049 \times 1 \approx 0.0000059049$

**Step 3:**

Now sum the probabilities $P(Y = 9) $ and $ P(Y = 10)$:

$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$

$P(Y \geq 9) = 0.000137781 + 0.0000059049$

$P(Y \geq 9) \approx 0.0001436859$

**CONCLUSION**

#### $\textbf{Task 3:}$ Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

---

$\textbf{Question 1: }$ *Find a general formula for the expected value of this distribution.*

***SOLUTION:***

Given the probability distribution for a Poisson random variable $Y$:

$p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$

Find the expected value $E[Y]$.

**Step 1:**

The expected value of $Y$ is given by:

$E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$

**Step 2:**

Substitute $p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $ into the formula:

$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

**Step 3:**

Notice that for $y = 0$, the term is zero because $y \cdot p(0) = 0$. Thus, can start the sum from $y = 1$:

$E[Y] = \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

**Step 4:**

Factor out the constant $e^{-\lambda}$:

$E[Y] = e^{-\lambda} \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!}$

**Step 5:**

Rewrite $y \cdot \frac{\lambda^y}{y!} $ as $ y \cdot \frac{\lambda \cdot \lambda^{y-1}}{y \cdot (y-1)!}$:

$y \cdot \frac{\lambda^y}{y!} = \lambda \cdot \frac{\lambda^{y-1}}{(y-1)!}$

Rewrite the sum:

$E[Y] = e^{-\lambda} \sum_{y=1}^{\infty} \lambda \cdot \frac{\lambda^{y-1}}{(y-1)!}$

**Step 6:**

Let $k = y - 1$. When $y = 1$, $k = 0$, and as $y \to \infty$, $k \to \infty$. 

Therefore:

$E[Y] = e^{-\lambda} \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!}$

**Step 7:**

The sum $\sum_{k=0}^{\infty} \frac{\lambda^k}{k!}$ is the Taylor series expansion for $e^\lambda$:

$E[Y] = e^{-\lambda} \lambda \cdot e^{\lambda}$

**Step 8:**

Simplify the product:

$E[Y] = \lambda \cdot e^{-\lambda} \cdot e^{\lambda} = \lambda$

$E[Y] = \lambda$

**CONCLUSION**

Thus, the expected value $E[Y]$ of the given Poisson distribution is:
$E[Y] = \lambda$
